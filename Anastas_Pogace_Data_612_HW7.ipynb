## About the assignment

In the industry, analytics professionals often explore data's measure variables grouped by one or multiple dimension variables. For example, the metrics of interest could be the average, max, min, or percentage (mix) of sales by month, location, etc.

The followings are the sample code snippet using the custom user functions together with `groupby()` and `apply()` in another way:


```
import pandas as pd

# the mpg data is used for demonstration
df = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv")
df.head()
```

```
def summarize_data(g):
  return(
    pd.Series({
      "avg_MPG": g.mpg.mean(),
      "min_MPG": g.mpg.min(),
      "max_MPG": g.mpg.max(),
      "sample_count": len(g),
      "mix": g.cylinders.sum()/df.cylinders.sum(),
    })
  )

df.groupby("cylinders").apply(summarize_data)
```

```
def Summarize_data_gby(g):
  return(
    pd.Series({
      "Hits_mean": g.H.mean(),
      "Doubles_mean": g['2B'].mean(),
      "Triples_mean": g['3B'].mean(),
      'HomeRun_mean': g.HR.mean(),
      'HomeRun_std' : g.HR.std(),
      'HomeRun_max' : g.HR.max(),
      'sample_count': len(g),
    })
  )

Batting_Info = df.groupby("yearID").apply(Summarize_data_gby)
Batting_Info.head(25)
```
